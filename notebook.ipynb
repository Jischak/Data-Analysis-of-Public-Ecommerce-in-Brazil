### Cleaning Data

```python
# check dataset
order_items_df.info()

# convert shipping_limit_date data type from object to datetime
order_items_df["shipping_limit_date"] = pd.to_datetime( order_items_df["shipping_limit_date"] )

# After convertion
print("\nAfter convertion")
order_items_df.info()

# show dataset
print("\nShow dataset")
order_items_df.head()
```

```python
# check dataset info
order_payments_df.info()

# convert payment_type data type from object to categorical
payment_type = pd.CategoricalDtype(order_payments_df["payment_type"].unique())
order_payments_df["payment_type"] = order_payments_df["payment_type"].astype(payment_type)

#change value boleto to barcode
order_payments_df["payment_type"] = order_payments_df["payment_type"].replace(["boleto"], "barcode")

# After convertion
print("\nAfter convertion")
order_payments_df.info()

# show dataset
print("\nShow dataset")
order_payments_df.head()
print(order_payments_df.loc[ order_payments_df["payment_type"] == "barcode" ])
```

```python
# check dataset info
orders_df.info()

# convert order_status data type from object to categorical
order_status = pd.CategoricalDtype(orders_df["order_status"].unique())
orders_df["order_status"] = orders_df["order_status"].astype(order_status)

# convert a few columns data type from object to datetime
orders_df["order_purchase_timestamp"] = pd.to_datetime( orders_df["order_purchase_timestamp"] )
orders_df["order_approved_at"] = pd.to_datetime( orders_df["order_approved_at"] )
orders_df["order_delivered_carrier_date"] = pd.to_datetime( orders_df["order_delivered_carrier_date"] )
orders_df["order_delivered_customer_date"] = pd.to_datetime( orders_df["order_delivered_customer_date"] )
orders_df["order_estimated_delivery_date"] = pd.to_datetime( orders_df["order_estimated_delivery_date"] )

# fill NA value
orders_df["order_approved_at"] = orders_df["order_approved_at"].fillna(method="ffill")
orders_df["order_delivered_carrier_date"] = orders_df["order_delivered_carrier_date"].fillna(method="ffill")
orders_df["order_delivered_customer_date"] = orders_df["order_delivered_customer_date"].fillna(method="ffill")


# After convertion
print("\nAfter convertion")
orders_df.info()

print("\nCheck Null Values")
print(orders_df.isna().sum())

# show dataset
print("\nShow dataset")
orders_df.head()
```

```python
# check dataset info
products_df.info()

# Remove NAs value from dataset
print("\n", products_df.isna().sum())
print(products_df.shape)
products_df = products_df.dropna(axis=0, how="any")

# After remove NAs value
print("\nCheck Null Values")
print(products_df.isna().sum())
print(products_df.shape)

# show dataset
print("\nShow dataset")
products_df.head()
```

## Exploratory Data Analysis (EDA)

### Explore on customers_dataset

```python
# check summary descriptive statistic of dataset
print("\nSummary Descriptive Statistic")
print(custs_df.describe())

print("\nTotal Number of customers")
print(custs_df["customer_id"].nunique())

print("\nShow dataset")
custs_df.head()
```

```python
# ANALYSIS: Top 10 Regions with the Highest Number of Customers
customers_based_on_regions = custs_df.groupby("customer_city")["customer_id"].nunique().reset_index().sort_values(by="customer_id",ascending=False)

print(customers_based_on_regions.head(10))
```

### Explore on orders_dataset

```python
# check summary descriptive statistic of dataset
print("\nSummary Descriptive Statistic")
print(orders_df.describe(datetime_is_numeric=True))

# Check status order
print("\nStatus order")
print(orders_df["order_status"].unique())

print("\nShow dataset")
orders_df.head()
```

```python
# Merge orders_df dataset with custs_df
orders_df_merge = pd.merge(orders_df, custs_df, on='customer_id', how = 'inner')
orders_df_merge.head()

# Drop columns that not used
columns_used_orders_df_merge = ["order_id", "customer_id", "order_status", "customer_city"]
columns_not_used_orders_df_merge = [x for x in orders_df_merge.columns if x not in columns_used_orders_df_merge ]
orders_df_merge_minimal = orders_df_merge.drop(columns_not_used_orders_df_merge, axis=1)

# Check the unavailable orders
print(orders_df_merge_minimal.loc[ orders_df_merge_minimal["order_status"] == "unavailable"])

orders_df_merge_minimal.head()
```

```python
# Filter success order = delivered
orders_df_merge_minimal_delivered = orders_df_merge_minimal.loc[ orders_df_merge_minimal["order_status"] == "delivered" ]

# ANALYSIS: Top 10 highest number of orders delivered by region
print("Top 10 orders by region")
top_10_order_by_region = orders_df_merge_minimal_delivered.groupby("customer_city")["order_id"].nunique().reset_index().sort_values(by="order_id", ascending=False)
print(top_10_order_by_region.head(10))

# ANALYSIS:  Top 10 lowest number of orders delivered by region
print("\nLowest 10 orders by region")
lowest_10_order_by_region = orders_df_merge_minimal_delivered.groupby("customer_city")["order_id"].nunique().reset_index().sort_values(by="order_id")
print(lowest_10_order_by_region.head(10))
```

```python
# Change the product name in the dataset to English
products_df_merge = pd.merge(products_df, product_cat, on='product_category_name', how = 'left')
print(products_df_merge.head())

# Check the dataset info
print("\nCheck dataset info")
print(products_df_merge.info())

# Check the NA value of the dataset
print("\nCheck NA value of dataset")
print(products_df_merge.isna().sum())
products_df_merge.loc[ products_df_merge["product_category_name_english"].isna() == True ]

# Fill NA value in product_category_name_english column
products_df_merge["product_category_name_english"] = products_df_merge["product_category_name_english"].fillna("unknown_category")

# Check the dataset after filling each NA value
print("\nCheck dataset info")
print(products_df_merge.isna().sum())

# Remove the columns is not used in dataset
print("List columns in dataset")
print(products_df_merge.columns)
columns_used = ["product_id", "product_category_name_english"]
columns_not_used = [x for x in products_df_merge.columns if x not in columns_used ]
products_df_merge_minimal = products_df_merge.drop(columns_not_used, axis=1)


# Show the new dataset
products_df_merge_minimal.head()
```

### Explore on order_payments_dataset

```python
# check summary descriptive statistic of dataset
print("\nSummary Descriptive Statistic")
print(order_payments_df.describe())

print("\nShow dataset")
order_payments_df.head()
```

```python
# Merge order_items_df_merge dataset with order_payments_df
order_payments_df_merge = pd.merge(order_items_df_merge, order_payments_df, on='order_id', how = 'inner')
order_payments_df_merge.head()
```

```python
# create new dataset that will export
main_data = pd.merge(order_payments_df_merge, orders_df_merge_minimal, on='order_id', how = 'inner')
main_data.head()
```

```python
# ANALYSIS: Top 10 Category Products with The Highest Amount of Orders
top_10_highest_order_PC = main_data.groupby("product_category_name_english")["order_id"].nunique().reset_index().sort_values("order_id", ascending=False)
print("Top 10 Category Products with The Highest Amount of Orders")
print(top_10_highest_order_PC.head(10))

print("\nTotal Number of orders")
print(top_10_highest_order_PC["order_id"].sum())

# ANALYSIS: Top 10 Category Products with The Lowest Amount of Orders
print("\nTop 10 Category Products with The Lowest Amount of Orders")
top_10_lowest_order_PC = main_data.groupby("product_category_name_english")["order_id"].nunique().reset_index().sort_values("order_id")
print(top_10_lowest_order_PC.head(10))
```

```python
print("\nTotal number of cities that placed an order")
print(main_data["customer_city"].nunique())

# ANALYSIS: Top 10 Number of Product Category Orders by Region
top_10_PC_by_region = main_data.groupby(["product_category_name_english", "customer_city"])["order_id"].nunique().reset_index("product_category_name_english").sort_values("order_id", ascending=False)
print("\nTop 10 Category Products with The Highest Amount of Orders")
top_10_PC_by_region.head(30)
```

```python
total_orders_by_payment_type = main_data.groupby("payment_type")["order_id"].nunique().reset_index().sort_values("order_id", ascending=False)
print("Total Numbers of orders based on Payment Type")
print(total_orders_by_payment_type.head())

# ANALYSIS: Top 10 Number of Product Category Orders by Payment Type
top_10_PC_by_PT = main_data.groupby(["product_category_name_english", "payment_type"])["order_id"].nunique().reset_index("product_category_name_english").sort_values("order_id", ascending=False)
print("\nTop 10 Number of Product Category Orders by Payment Type")
top_10_PC_by_PT.head(30)
```
